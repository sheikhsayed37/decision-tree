````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split


from sklearn.datasets import load_iris

from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,precision_score,recall_score

from sklearn.metrics import confusion_matrix,classification_report

from sklearn.model_selection import GridSearchCV
````

````python
df=load_iris(as_frame=True).frame

df
````

````python
x = df.drop(['target'],axis=1)
y = df['target']
x_train , x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42,stratify=y)
````

````python
model=DecisionTreeClassifier()
model.fit(x_train,y_train)
y_pred= model.predict(x_test)
y_pred
````

````python
print(classification_report(y_test,y_pred))
````

````python
grid ={
    "criterion":['gini','entropy'],
    'max_depth':[2,3,4,5,6,7,8,None],
}

grid_search=GridSearchCV(
    estimator=model,
    param_grid=grid,
    cv=5
    )
grid_search.fit(x_train,y_train)
````

````python
y_pred = grid_search.predict(x_test)

print(classification_report(y_test,y_pred))
````

#short note :

**Gini Impurity** মানে হলো—

> **একটি Node-এর Data কতটা মিশ্রিত (Impure) আছে, তার পরিমাপ।**

অথবা আরও সহজভাবে,

> **একটি Node থেকে Random একটি Sample নিলে সেটিকে ভুল Classify হওয়ার সম্ভাবনা কত?**

---

## উদাহরণ ১: Pure Node

ধরো একটি Node-এ আছে

```text
10 টি ফুল

Setosa = 10
Versicolor = 0
Virginica = 0
```

সবগুলো একই Class।

এখানে

**Gini = 0**

মানে **একদম Pure**।

---

## উদাহরণ ২: Mixed Node

ধরো

```text
10 টি ফুল

Setosa = 5
Versicolor = 5
```

এখানে দুইটা Class মিশে আছে।

**Gini = 0.5**

মানে Node Impure।

---

## উদাহরণ ৩

ধরো

```text
10 টি ফুল

Setosa = 4
Versicolor = 3
Virginica = 3
```

এখানে আরও বেশি Mixing।

Gini আরও বেশি হবে।

---

## Formula

[
\text{Gini} = 1 - \sum p_i^2
]

এখানে,

* (p_i) = প্রতিটি Class-এর Probability

---

### ছোট Example

ধরো

```text
Setosa = 8
Versicolor = 2
```

Probability

```text
Setosa = 8/10 = 0.8

Versicolor = 2/10 = 0.2
```

তাহলে

[
\text{Gini}
===========

1-(0.8^2+0.2^2)
]

[
=1-(0.64+0.04)
]

[
=0.32
]

অর্থাৎ **Gini = 0.32**।

---

## Decision Tree এটা কীভাবে ব্যবহার করে?

ধরো দুটি Feature আছে।

### Split 1

```text
Gini = 0.42
```

### Split 2

```text
Gini = 0.12
```

Decision Tree বলবে,

👉 **Split 2 ভালো**, কারণ Gini কম মানে Node বেশি Pure।

---

## সহজে মনে রাখো

* **Gini = 0** → একদম Pure (সব একই Class) ✅
* **Gini বেশি** → Class গুলো বেশি মিশে আছে ❌

---

### ১ লাইনে Viva Answer

**"Gini Impurity is a measure of how mixed the classes are in a node. A lower Gini value indicates a purer node, so Decision Tree chooses the split that minimizes Gini Impurity."**


# Entropy note

অবশ্যই। **Entropy**-কে Gini-এর মতোই সহজভাবে বুঝি।

---

# Entropy মানে কী?

**Entropy** হলো—

> **একটি Node-এর Data কতটা বিশৃঙ্খল (Disordered) বা Mixed আছে, তার পরিমাপ।**

অর্থাৎ,

* **Entropy কম = Data পরিষ্কার (Pure)**
* **Entropy বেশি = Data মিশ্রিত (Impure)**

---

## উদাহরণ ১: Pure Node

ধরো একটি Node-এ আছে

```text
10 টি ফুল

Setosa = 10
Versicolor = 0
Virginica = 0
```

সব একই Class।

তাহলে

```text
Entropy = 0
```

মানে **কোনো Confusion নেই।**

---

## উদাহরণ ২: Mixed Node

ধরো

```text
10 টি ফুল

Setosa = 5
Versicolor = 5
```

এখন দুইটা Class সমান।

তাহলে

```text
Entropy = 1
```

মানে **Confusion সবচেয়ে বেশি।**

---

## উদাহরণ ৩

ধরো

```text
10 টি ফুল

Setosa = 8
Versicolor = 2
```

এখানে

Data পুরোপুরি Mixed না।

তাই

```text
Entropy ≈ 0.72
```

অর্থাৎ কিছুটা Confusion আছে।

---

# Formula

[
Entropy=-\sum p_i\log_2(p_i)
]

এখানে

* (p_i) = প্রতিটি Class-এর Probability

---

### Example

ধরো

```text
Setosa = 8
Versicolor = 2
```

Probability

```text
Setosa = 0.8

Versicolor = 0.2
```

তাহলে

[
Entropy
=======

-(0.8\log_2 0.8+0.2\log_2 0.2)
]

[
≈0.72
]

---

# Decision Tree কীভাবে ব্যবহার করে?

ধরো দুটি Split পাওয়া গেল।

### Split 1

```text
Entropy = 0.60
```

### Split 2

```text
Entropy = 0.15
```

Decision Tree বলবে

👉 **Split 2 ভালো।**

কারণ

**Entropy কম = Node বেশি Pure।**

---

# তাহলে Information Gain কোথায় আসে?

Decision Tree আগে দেখে

Split করার আগে Entropy কত।

ধরো

```text
Before Split

Entropy = 1
```

Split করার পরে

```text
After Split

Entropy = 0.2
```

তাহলে

```text
Information Gain

= 1 - 0.2

= 0.8
```

অর্থাৎ

**Split করার ফলে Confusion অনেক কমে গেছে।**

তাই Decision Tree এই Split-টাই বেছে নেবে।

---



## খুব সহজ ট্রিক

ধরো একটা ক্লাসে ১০ জন ছাত্র আছে।

### সবাই একই সেকশনের

```text
A A A A A A A A A A
```

👉 কোনো Confusion নেই।

**Entropy = 0**

---

### অর্ধেক A, অর্ধেক B

```text
A A A A A
B B B B B
```

👉 কে কোন সেকশনের, বুঝতে Confusion বেশি।

**Entropy = সবচেয়ে বেশি (১)**

---

### Viva-তে ১ লাইনের উত্তর

> **Entropy is a measure of the uncertainty or randomness in a node. Lower entropy means the node is purer, while higher entropy means the classes are more mixed.**


# Gini vs Entropy

| Gini                                | Entropy                                                               |
| ----------------------------------- | --------------------------------------------------------------------- |
| Data কতটা Mixed তা মাপে             | Data কতটা Confused তা মাপে                                            |
| Gini কম হলে ভালো                    | Entropy কম হলে ভালো                                                   |
| Gini ব্যবহার করে Split নির্বাচন করে | Information Gain (Entropy কমার পরিমাণ) ব্যবহার করে Split নির্বাচন করে |
