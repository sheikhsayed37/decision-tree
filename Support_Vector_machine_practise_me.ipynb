````python
import numpy as np
import pandas as pd


from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler


from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score

````

সহজ বাংলায়:

### ১. Accuracy (একিউরেসি)

**অর্থ:** মডেল মোট ভবিষ্যদ্বাণীর মধ্যে কতগুলো সঠিকভাবে করেছে।

**সূত্র:**

[
\text{Accuracy}=\frac{TP+TN}{TP+TN+FP+FN}
]

**সহজভাবে:**
মডেলের **সামগ্রিক সঠিকতার হার**।

---

### ২. Precision (প্রিসিশন)

**অর্থ:** মডেল যেগুলোকে **Positive (Malignant)** বলেছে, তার মধ্যে আসলে কতগুলো সত্যিই Positive।

**সূত্র:**

[
\text{Precision}=\frac{TP}{TP+FP}
]

**সহজভাবে:**
মডেল **Positive বললে সেটি কতটা বিশ্বাসযোগ্য**, তা বোঝায়।

---

### ৩. Recall (রিকল)

**অর্থ:** আসল **Positive (Malignant)** রোগীদের মধ্যে মডেল কতজনকে সঠিকভাবে শনাক্ত করতে পেরেছে।

**সূত্র:**

[
\text{Recall}=\frac{TP}{TP+FN}
]

**সহজভাবে:**
মডেল **আসল Positive কেসগুলোর কত শতাংশ ধরতে পেরেছে**, তা বোঝায়।

---


* **Accuracy:** মডেল মোট কত শতাংশ সঠিক Prediction করেছে।
* **Precision:** মডেল যাদের Positive বলেছে, তাদের মধ্যে কত শতাংশ সত্যিই Positive।
* **Recall:** আসল Positive কেসগুলোর মধ্যে মডেল কত শতাংশ সঠিকভাবে শনাক্ত করেছে।

Dataset: https://www.kaggle.com/datasets/fedesoriano/heart-failure-prediction
Objective
Predict whether a patient has heart disease.

Tasks

Train four SVM models:

    Linear
    Polynomial
    RBF
    Sigmoid

**Compare**

Accuracy

Precision

Recall


````python
df=pd.read_csv('heart.csv')
df.sample(12)
````

````python
df.isnull().sum()
````

````python
X = df.drop("HeartDisease", axis=1)
y = df["HeartDisease"]
````

````python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)
````

````python
categorical_columns = X.select_dtypes(include="object").columns

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

preprocessor = ColumnTransformer(
    transformers=[
        (
          "cat",OneHotEncoder(handle_unknown="ignore"),categorical_columns
        )
    ],
    remainder="passthrough"
)
````

````python
X_train_encoded = preprocessor.fit_transform(X_train)
````

````python
from sklearn.pipeline import Pipeline
from sklearn.svm import SVC

SVC_model = Pipeline(
    steps=[
        ("preprocessor", preprocessor),
        ("model", SVC())
    ]
)
````

````python
grid_param = [
    {
        "model__kernel": ["linear"],
        "model__C": [0.01, 0.1, 1, 10, 50, 100]
    },
    {
        "model__kernel": ["poly"],
        "model__C": [0.01, 0.1, 1, 10, 100],
        "model__degree": [2, 3]
    },
    {
        "model__kernel": ["rbf"],
        "model__C": [0.01, 0.1, 1, 10, 100],
        "model__gamma": [0.01, 0.1, 5, 10, "scale", "auto"]
    },
    {
        "model__kernel": ["sigmoid"],
        "model__C": [0.01, 0.1, 1, 10, 100],
        "model__gamma": [0.01, 0.1, "scale", "auto"]
    }
]
````

````python
from sklearn.model_selection import GridSearchCV

best_SVC_model = GridSearchCV(
    estimator=SVC_model,
    param_grid=grid_param,
    cv=5,
    scoring="accuracy"
)
best_SVC_model.fit(X_train, y_train)
````

````python
print(best_SVC_model.best_params_)
print(best_SVC_model.best_score_)
````

````python
y_pred = best_SVC_model.predict(X_test)


from sklearn.metrics import accuracy_score, precision_score, recall_score


print("Accuracy :", accuracy_score(y_test, y_pred))
print("Precision :", precision_score(y_test, y_pred))
print("Recall :", recall_score(y_test, y_pred))
````


#important note

 এই `grid_param`-টাই হলো **GridSearchCV-এর সবচেয়ে গুরুত্বপূর্ণ অংশ**। এখানে Model-কে বলছো **কোন কোন Parameter-এর Combination পরীক্ষা করতে হবে**।

তবে আগে একটা ছোট ভুল আছে। তুমি `;` (semicolon) ব্যবহার করেছ, Python-এ এখানে `,` (comma) হবে।

সঠিক কোড:

```python
grid_param = [
    {
        "model__kernel": ["linear"],
        "model__C": [0.01, 0.1, 1, 2, 3, 4, 5, 10]
    },
    {
        "model__kernel": ["rbf"],
        "model__C": [0.01, 0.1, 1, 2, 3, 4, 5, 10],
        "model__gamma": [0.01, 0.1, "scale", "auto"]
    },
    {
        "model__kernel": ["poly"],
        "model__C": [0.01, 0.1, 1, 2, 3, 4, 5, 10],
        "model__degree": [2, 3, 4, 6, 7]
    }
]
```

---

## `grid_param = []` কেন List?

কারণ তুমি **একাধিক Kernel** পরীক্ষা করছো।

* ১ম Dictionary → Linear
* ২য় Dictionary → RBF
* ৩য় Dictionary → Polynomial

---

## ১ম Dictionary

```python
{
    "model__kernel": ["linear"],
    "model__C": [0.01, 0.1, 1, 2, 3, 4, 5, 10]
}
```

এর মানে:

GridSearchCV নিচের Model-গুলো চালাবে:

```text
Linear + C = 0.01
Linear + C = 0.1
Linear + C = 1
Linear + C = 2
...
Linear + C = 10
```

মোট **৮টি Model**।

---

## `C` কী?

`C` হলো **Regularization Parameter**।

* **C ছোট** (0.01, 0.1)

  * Model সহজ (Simple)
  * কিছু ভুল মেনে নেয়
  * Overfitting কম হয়

* **C বড়** (10, 100)

  * ভুল কম করতে চায়
  * Training Data ভালোভাবে Fit করে
  * Overfitting হতে পারে

---

## ২য় Dictionary (RBF)

```python
{
    "model__kernel": ["rbf"],
    "model__C": [...],
    "model__gamma": [...]
}
```

এখানে GridSearchCV **C** এবং **gamma**-এর সব Combination পরীক্ষা করবে।

উদাহরণ:

```text
RBF + C=0.01 + gamma=0.01
RBF + C=0.01 + gamma=0.1
RBF + C=0.01 + gamma=scale
RBF + C=0.01 + gamma=auto

RBF + C=0.1 + gamma=0.01
...
```

যদি

* `C` = 8টি মান
* `gamma` = 4টি মান

তাহলে

**8 × 4 = 32টি Model** ট্রাই করবে।

---

## `gamma` কী?

`gamma` শুধু **RBF**, **Poly**, **Sigmoid** Kernel-এ কাজ করে।

* **gamma ছোট** → দূরের Point-ও প্রভাব ফেলে → Boundary মসৃণ।
* **gamma বড়** → কাছের Point বেশি প্রভাব ফেলে → Boundary জটিল হতে পারে।

---

## ৩য় Dictionary (Polynomial)

```python
{
    "model__kernel": ["poly"],
    "model__C": [...],
    "model__degree": [2,3,4,6,7]
}
```

এখানে

* `degree = 2` → Quadratic
* `degree = 3` → Cubic
* `degree = 4` → Fourth-degree polynomial

যদি

* `C` = 8টি
* `degree` = 5টি

তাহলে

**8 × 5 = 40টি Model** পরীক্ষা করবে।

---

## GridSearchCV কীভাবে Best Model বের করে?

ধরো সে মোট **80টি Model** ট্রেন করল।

তারপর প্রতিটির **Cross Validation Score** বের করবে।

যেটার Score সবচেয়ে বেশি, সেটাই হবে:

```python
best_SVC_model.best_params_
```

যেমন:

```python
{'model__kernel': 'linear',
 'model__C': 10}
```

অর্থাৎ GridSearchCV বলছে:

> **সব Combination-এর মধ্যে `Linear Kernel` এবং `C = 10` সবচেয়ে ভালো ফল দিয়েছে।**

এটাই `best_params_`-এর অর্থ।






# ১. `degree` কী? (শুধু Polynomial Kernel-এর জন্য)

`degree` মানে **Curve কতটা বাঁকা (complex) হবে**।

* `degree = 2` → একটু বাঁকা
* `degree = 3` → আরও বাঁকা
* `degree = 4` → আরও জটিল
* `degree` যত বাড়বে → Model তত জটিল Pattern শিখতে পারবে।

**সহজ উদাহরণ:**

* `degree = 2` → 😊 (হালকা বাঁক)
* `degree = 7` → 🌀 (অনেক বেশি বাঁক)

যদি Data খুব জটিল হয়, বড় `degree` কাজে আসতে পারে। কিন্তু খুব বড় `degree` দিলে **Overfitting** হতে পারে।

---

# ২. `gamma` কী? (RBF, Poly, Sigmoid-এর জন্য)

`gamma` ঠিক করে **একটি Data Point তার আশেপাশে কতটা প্রভাব ফেলবে**।

### `gamma` ছোট

```text
gamma = 0.01
```

একটি Point দূরের Point-কেও প্রভাবিত করে।

➡️ Boundary মসৃণ (Smooth) হয়।

---

### `gamma` বড়

```text
gamma = 10
```

একটি Point শুধু খুব কাছের Point-কে প্রভাবিত করে।

➡️ Boundary অনেক জটিল (Complex) হয়।

---

## সহজ ট্রিক

### `degree`

> **degree = Curve-এর বাঁক কতটা হবে।**

* ছোট degree → সহজ Curve
* বড় degree → জটিল Curve

---

### `gamma`

> **gamma = একটি Point-এর প্রভাব কত দূর পর্যন্ত যাবে।**

* ছোট gamma → প্রভাব দূর পর্যন্ত যায়।
* বড় gamma → প্রভাব শুধু কাছাকাছি থাকে।

---

## পরীক্ষার জন্য এক লাইনে

* **Degree:** Polynomial Kernel-এর Curve-এর জটিলতা নির্ধারণ করে।
* **Gamma:** একটি Training Point-এর প্রভাব কত দূর পর্যন্ত থাকবে তা নির্ধারণ করে।

এভাবেই লিখলে পরীক্ষায়ও নম্বর পাওয়া যাবে এবং GridSearchCV-তে `degree` ও `gamma` কেন টিউন করা হয় সেটাও বুঝতে সুবিধা হবে।
